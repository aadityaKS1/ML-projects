```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://merojob.com/category/it-telecommunication/?page=2'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
webpage=requests.get(url,headers=headers).text
print(webpage)
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
cname=soup.find_all('h3')
[i.get_text(strip=True) for i in cname]

post=soup.find_all('h1',class_='text-primary font-weight-bold media-heading h4')
post_list=[i.get_text(strip=True) for i in post]
print(post_list)
```

```python
skills=soup.find_all('span',itemprop="skills")
skills_list=[i.get_text(strip=True) for i in skills ]
skill_container=skills[0]
specific_skill=[i.get_text(strip=True) for i in skill_container.find_all('span') ]

print(post_list[0])
print(skills_list[0])
print(specific_skill[0])
    
```
